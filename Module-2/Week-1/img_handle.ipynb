```python
import numpy as np

def rgb_to_luminosity_grayscale(img_matrix):
    # Define the weights for the luminosity method
    weights = [0.21, 0.72, 0.07]
    
    # Calculate the grayscale values using the weights
    grayscale_matrix = np.dot(img_matrix[...,:3], weights)
    
    return grayscale_matrix.astype('uint8')

gray_img = rgb_to_luminosity_grayscale(img_rgb)

plt.imshow(gray_img)

gray_img[0,0]
```
